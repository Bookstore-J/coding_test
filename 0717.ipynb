```python
import sys

input = sys.stdin.readline

n,m,k = map(int, input().split())

from collections import deque

graph = [[0]*(m) for i in range(n)]


for _ in range(k):
    a,b = map(int, input().split())
    graph[a-1][b-1] = 1
    
def bfs(x,y):
    cnt = 1
    queue = deque([(x,y)])
    graph[x][y] = 0
    
    dx = [0,0,1,-1]
    dy = [1,-1,0,0]
    
    while queue:
        x,y = queue.popleft()
        
        for _ in range(4):
            nx = x + dx[_]
            ny = y + dy[_]
            
            if 0<=nx<n and 0<=ny<m and graph[nx][ny] == 1:
                queue.append((nx,ny))
                graph[nx][ny] = 0 
                cnt += 1
                
    return cnt
    
    

total_list = []
for i in range(n):
    for j in range(m):
        if graph[i][j] == 1:
            total_list.append(bfs(i,j))

max(total_list)
```

```python
## 10541

N = int(input())

for i in range(N):
    n = int(input())

    lst = [[] for i in range(n+1)]
    lst_ = list(map(int, input().split(' ')))


    for _ in range(n):
        lst[_+1].append(lst_[_])


    visited = [False for _ in range(n+1)]

    def dfs(v):

        visited[v] = True    
        for i in lst[v]:        
            if not visited[i]:
                dfs(i)

    cnt = 0
    for i in range(1, n+1):
        if visited[i] == False:
            dfs(i)
            cnt += 1

    print(cnt)
```
