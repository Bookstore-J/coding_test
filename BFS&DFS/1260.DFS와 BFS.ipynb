## 1260. DFS와 BFS

그래프를 DFS로 탐색한 결과와 BFS로 탐색한 결과를 출력하는 프로그램을 작성하시오.    
단, 방문할 수 있는 정점이 여러 개인 경우에는 정점 번호가 작은 것을 먼저 방문하고, 더 이상 방문할 수 있는 점이 없는 경우 종료한다.    
정점 번호는 1번부터 N번까지이다.   

```python
n,m,v = map(int, input().split(' '))
```

```python
graph = [[0 for i in range(n+1)] for i in range(n+1)]
```

```python
visited = [[0 for i in range(n+1)] for i in range(n+1)]
```

```python
for _ in range(m):
    a,b = map(int, input().split(' '))
    graph[a][b]  = graph[b][a] = 1
```

```python
def dfs(v):
    for i in graph[v]:
        if i == 1 and visited[v][i] == 0:
            visited[v][i] == 1
            print(v, end= ' ')
            dfs(i)
```

```python
dfs(v)
```

```python
from collections import deque

n,m,v = map(int, input().split())

graph =  [[0] * (n+1) for _ in range(n+1)]
# [0] * (n) = [0, 0, 0, 0, ... 0] (n개) 
for i in range(m):
    a,b = map(int, input().split())
    graph[a][b] = graph[b][a] = 1  # 인접행렬 그래프 표현 // 인접리스트도 있다.

# 노드 중심으로 방문처리 
visited1 = [0] * (n+1)
visited2 = [0] * (n+1)

def dfs(v):
    visited1[v] = 1
    print(v, end=' ')
    for i in range(1, n+1):
        if graph[v][i] == 1 and visited1[i] == 0:
            dfs(i)

def bfs(v):
    queue = deque([v])
    visited2[v] = 1
    while queue:
        v = queue.popleft()
        print(v, end=' ')
        for i in range(1, n+1):
            if (visited2[i] == 0 and graph[v][i] == 1):
                queue.append(i)
                visited2[i] = 1
```

```python
dfs(v)
print()
bfs(v)
```
